## 1. How many are there in each Team and the percentage splitting with respect to the total employees.


```python
df1.columns
```

```python
df1.Team.unique()
```

```python
df1.Team.nunique()
```

```python
df1['Team'].value_counts()
```

```python
team_count=(df1['Team'].value_counts()/458)*100
team_count
```

```python

plt.plot(team_count.index,team_count.values)
plt.xticks(rotation='vertical')
plt.title('Employees vs Teams',fontdict={'fontweight': 'bold'})
plt.xlabel('Teams')
plt.ylabel('Percentage of splitting')
plt.show()


team_count=(df1['Team'].value_counts()/458)*100
plt.figure(figsize=(11,11))
plt.pie(team_count.values,labels=team_count.index,autopct='%1.1f%%')
plt.title('Employees vs Teams',fontdict={'fontweight': 'bold'})
plt.show()
```

#### Data Story-
From analysing the above bar graph , it's clear that the team "New Orleans Pelicans" has more percentage split compared to other teams , Also there is a identical pattern for some teams starting from "Brooklyn Nets" to "Dallas Mavericks".


## 2.Segregate the employees w.r.t different positions.


```python
position=df1['Position'].unique()
position
```

```python
positions=df1['Position'].value_counts()
positions
```

```python
plt.bar(positions.index,positions.values,color='b',width=.7,edgecolor='black')
plt.xticks(rotation='vertical')
plt.xlabel('Position')
plt.ylabel('No. of Employees')
plt.title('Employees vs Teams',fontdict={'fontweight': 'bold'})
plt.show()

label=df1['Position'].value_counts()
plt.pie(label.values,labels=label.index,autopct='%1.1f%%')
plt.title('Position Ratio',color='b',fontdict={'fontweight': 'bold'})
plt.show()
```

## 3.Find from which age group most of the employees belong to.

```python
age_grp=df1['Age'].value_counts()
age_grp.sort_values(ascending=True)
```

```python
plt.hist(df1['Age'],color='y',edgecolor='b')
plt.xlabel('Age')
plt.ylabel('Freguency of aAge')
plt.title('Age Distribution of Employees',color='red',fontdict={'fontweight': 'bold'})
plt.show()
plt.scatter(age_grp.index,age_grp.values,)
plt.xlabel('Age')
plt.ylabel('Freguency')
plt.ylim(0,50)
plt.xlim(15,42)
plt.grid()
plt.title('Age Frequency',color='g',fontdict={'fontweight': 'bold'})
plt.show()
```

#### Data Story-
 From the above graphs we can understand that, most of the employees are in the age group 20-30, and there is less number of employees in the age group of 40. Also there are number of employees of age 15-20 which is between both 20-30.

## 4.Find out under which team and position, spending in terms of salary is high.

```python
team_total_salary=df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
team_total_salary
```

```python
team_total_salary.idxmax()
```

```python
plt.figure(figsize=(12,6))
clr=sns.color_palette('cubehelix')
plt.bar(team_total_salary.index,team_total_salary.values,width=.8,edgecolor='black',color=clr)
plt.xticks(rotation='vertical')
plt.xlabel('Teams',fontdict={'fontweight': 'bold'})
plt.ylabel('Total Salary',fontdict={'fontweight': 'bold'})
plt.title(' Teams vs Salary',color='r',fontdict={'fontweight': 'bold'})
plt.show()
```

### Data Story-
From the above graph figure we can understand that 'Cleveland Cavaliers' is the Team spend most in terms of Salary. 

```python
position_total_salary=df.groupby('Position')['Salary'].sum()
position_total_salary
```

```python
plt.figure(figsize=(12,6))
clr=sns.color_palette('cubehelix')
plt.bar(position_total_salary.index,position_total_salary.values,width=.8,edgecolor='black',color=clr)
plt.xlabel('Positions',fontdict={'fontweight': 'bold'})
plt.ylabel('Total Salary',fontdict={'fontweight': 'bold'})
plt.title(' Positions vs Salary',color='r',fontdict={'fontweight': 'bold'})
plt.show()
```

### Data Story -
 The Maximum Salary spend for Position 'C'

## 5..Find if there is any correlation between Age and Salary, Represent it visually.

```python
corr_matrix= df1[['Age','Salary']].corr()
corr_matrix
```

### Visual Representation


```python
sns.heatmap(corr_matrix,annot=True,cmap='viridis')
plt.title('Correlation matrix of Age and Salary',color='g',fontdict={'fontweight': 'bold'})
plt.show()
```

#### Data Story-
Map Shows that Age and Salary are Postivily Correlation ,But the bond is not very strong
